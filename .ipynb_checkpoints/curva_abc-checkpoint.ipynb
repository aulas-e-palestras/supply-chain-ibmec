## Importações

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

## Leitura dos dados

```python
data = (
    pd.read_csv('data/sales_retail.csv',usecols=['product_id','store_id','date','sales','stock','price'])
    .assign(date = lambda x: pd.to_datetime(x.date))
    .dropna(subset=['sales','stock','price'],how='any')
    .assign(total_sale = lambda x: x.sales * x.price)
    
)
```

## Análise dos dados

### Curva ABC

Iremos analisar somente os itens A para o trabalho visto sua capacidade de impacto no resultado financeiro da empresa.

```python
curva_abc = (
    data
    .groupby('product_id')
    .sum()[['total_sale']]
    .sort_values('total_sale')
    .assign(percent_curva = lambda x: x.total_sale / x.total_sale.sum())
    .sort_values('percent_curva',ascending=False)
)
```

```python
ids = []
percent = 0
i = 1
done = False

while not done:

    for idx in range(0,i):
        if curva_abc.iloc[:idx,1].sum() <= 0.8:
            i += 1
            done = False
        else:
            percent = curva_abc.iloc[:idx,1].sum()
            ids.extend(curva_abc.iloc[:idx,1].index.values.tolist())
            done = True
            break
```

```python
curva_abc_plot = (
    curva_abc[['percent_curva']]
    .cumsum()
    .assign(item_percent = lambda x: np.arange(1,len(x)+1))
    .assign(item_percent = lambda x: x.item_percent / len(x))
    
)
```

```python
f, ax = plt.subplots(figsize=(16,9))

ax.plot(curva_abc_plot.set_index('item_percent').percent_curva)
ax.axvspan(xmin=0,
           xmax=curva_abc_plot.loc[ids[-1],'item_percent'],
           ymax=percent,
           color='tab:red',
           label=f'receita : {round(percent,3)}\nitens : {round(curva_abc_plot.loc[ids[-1],"item_percent"],3)}',
           alpha=0.2)

ax.set_title('Curva ABC')
ax.set_xlabel('porcentagem de itens')
ax.set_ylabel('porcentagem da receita')
plt.legend()
plt.show()
```

```python
del curva_abc
del curva_abc_plot
```

### Estoque e demanda dos SKUs A

```python
data_a = (
    data[data.product_id.isin(ids)]
    .reset_index(drop=True)
    .query("date >= '2019-01-01'")
    .assign(mes = lambda x: x.date.dt.month)
    .assign(semana = lambda x: x.date.dt.isocalendar().week)
)

data_a
```

```python
data_plot = (
    data_a
    .pivot_table(index='product_id',columns='semana',values='sales',aggfunc='sum',fill_value=0)
    .T
    .rolling(2)
    .mean()
)

data_plot
```

```python
plot_data = data_a.pivot_table(index='product_id',columns='date',values='sales',aggfunc='sum',fill_value=0)
```

```python
plot_data.T.rolling(30)
```

```python
f, ax = plt.subplots(figsize=(16,9))

ax.plot(plot_data.T,color='tab:blue',alpha=0.6)
plt.show()
```
