# Testando hipóteses de produção

```python
def testar_hipotese_produção(vinho,cerveja):
    """
    Função que testa hipoteses de produção
    e retorna os valores de saldo de capacidade produtiva, estoque
    e lucro total associado a quantidade de lotes de vinho e cerveja.
    """
    
    capacidade_de_producao = 60
    estoque = 110
    lucro = 0
    
    capacidade_de_producao -= vinho * 9 + cerveja * 5
    estoque -= vinho * 10 + cerveja * 20
    lucro += vinho * 5 + cerveja * 4.5
    
    return (capacidade_de_producao, estoque, lucro)
```

# Solucionando com PuLP

```python
## definindo o problema

model = LpProblem('sequenciamento_xpto',LpMaximize)
```

```python
## definindo as variáveis de decisão

vinho = LpVariable('vinho', lowBound=0, upBound=None, cat='Integer')
cerveja = LpVariable('cerveja', lowBound=0, upBound=None, cat='Integer')
```

```python
## definindo a função objetivo

model += 5 * vinho + 4.5 * cerveja
```

```python
## definido as restrições

model += 9 * vinho + 5 * cerveja <= 60
model += 10 * vinho + 20 * cerveja <= 110
model += vinho <= 6
```

```python
## solucionando o problema

resolucao = model.solve()
```
